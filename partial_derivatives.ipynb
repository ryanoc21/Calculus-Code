```python
import sympy as sym 
from IPython.display import display, Math
```

```python
from sympy.abc import x,y

f = x**2 + x*y**2

display(Math('\\frac{\\partial f}{\\partial x } = %s' %(sym.latex(sym.diff(f,x)))))

display(Math('\\frac{\\partial f}{\\partial y } = %s' %(sym.latex(sym.diff(f,y)))))
```

```python
#EXERICISE, variables are the same as in the cell above 

p = sym.plotting.plot3d(f,(x,-3,3),(y,-3,3), title = '$f(x,y) = %s $'%sym.latex(f))
pxdiff = sym.plotting.plot3d(sym.diff(f,x),(x,-3,3),(y,-3,3),\
                            title = '$\\frac{\partial f}{\partial x} = %s $'%sym.latex(sym.diff(f,x)))
pydiff = sym.plotting.plot3d(sym.diff(f,y),(x,-3,3),(y,-3,3),\
                            title = '$\\frac{\partial f}{\partial y} = %s $'%sym.latex(sym.diff(f,y)))
```
