```python
import sympy as sym 
from IPython.display import display, Math
```

```python
#COMPUTING THE INDEFINITE INTEGRAL 
x = sym.symbols('x') 
fx = x 
sym.integrate(fx)
```

```python
#COMPUTING THE DEFINITE INTEGRAL 
sym.integrate(fx,(x,0,1))
#in the tuple,x is the variable and the two numbers are the range you integrate over 
```

```python
#Plotting the function 
display(Math('\\text{indefinite integral} = %s'%(sym.latex(sym.integrate(fx)))))
display(Math('\\text{definite integral} = %s'%(sym.latex(sym.integrate(fx,(x,0,1))))))


#Plotting the function 
p = sym.plotting.plot(fx,show = False) 

#Setting plot boundaries 
p.xlim = [0,1]
p.ylim = [0,1]
p.show()
```

```python
f = x**3 /(x-2)

intf = sym.integrate(f) 
p = sym.plotting.plot(f,show = False )
p.extend(sym.plotting.plot(intf,(x,2.1,10), show = False, line_color = 'r'))

p[0].label = '$f(x) = %s $'%(sym.latex(f))
p[1].label = '$\\int f(x)dx = %s $'%(sym.latex(intf))
p.legend = True 
p.ylim = [-200,200]
p.show()
```

```python
#EXERCISE: FUNDAMENTAL THEOREM OF CALCULUS 

#The function 
fx = 2*x**3 + sym.sin(x)

#The integral of the derivative of the function
intfx= sym.integrate(sym.diff(fx))

display(Math('%s \\quad = \\quad %s'%(sym.latex(fx),sym.latex(intfx))))
```
