```python
import sympy as sym 
import numpy as np 
import matplotlib.pyplot as plt 
from IPython.display import display, Math
```

```python
#Example 1 
x = sym.symbols('x') 

fx = x**2
#To calculate the derivatice using sympy you use sym.diff
sym.diff(fx)
```

```python
#Example 2: using other numbers
x = sym.symbols('x')

for fx in range(1,4): 
    function = fx*x**(fx)
    deriv = sym.diff(function)
    display(Math('\\frac{df}{dx}%s = %s'%(sym.latex(function), sym.latex(deriv))))
    print(' ')
    
```

```python
#VARIOUS NOTATIONS USED TO DISPLAY THE DERIVATIVE 
fx = x**2
dfx = sym.diff(fx)

#Leibniz Notation
display(Math('\\text{Leibniz Notation}'))
display(Math('f(x) = %s, \\quad \\frac{df}{dx} = %s'%(sym.latex(fx), sym.latex(dfx))))

print('')

#Lagrange Notation 
display(Math('\\text{Lagrange Notation}'))
display(Math('f(x) = %s, \\quad f\' = %s'%(sym.latex(fx), sym.latex(dfx))))

print('')

#Newton Notation
display(Math('\\text{Newton Notation}'))
display(Math('f(x) = %s, \\quad \\dot{f} = %s'%(sym.latex(fx), sym.latex(dfx))))
```

```python
#PLOTTING THE DERIVATIVE

#To plot in sympy: 
import sympy.plotting.plot as symplot

#Function
fx = 3 - x**3

#PLOT 
#1.Plotting the function
plot = symplot(fx,(x,-3,3),show=False) #Note, the second input x is the range, same as np.linspace

#2.Plotting the derivative onto the function
plot.extend(symplot(sym.diff(fx),show = False))

#Making everything look pretty 
plot[1].line_color = 'r'#changes the color of the line, in this case the second plot (derivative)

plot[0].label = '$ f(x) = %s $' %sym.latex(fx) #labels the function 
plot[1].label = '$ f(x) = %s $' %sym.latex(sym.diff(fx)) #labels the derivative
plot.legend = True
plot.ylim = [-10,10]
plot.xlim = [-3,3]
plot.show()
```

```python
#EXERCISE 1: DEMONSTRATE THE "PRODUCT RULE" AND "SUMMATION RULE" OF DERIVATIVES 
x = sym.symbols('x') 

#Functions f(x) and g(x) 
fx = 3 + 2*x - 5*x**2 + 7*x**4
fdiv = sym.diff(fx) 

gx = 4*x**2 + x**5 
gdiv = sym.diff(gx) 

fbyg = fx*gx
fplusg = fx+gx
fdiv_xg = fdiv*gx
gdiv_xf = gdiv*fx

#SUMMATION RULE
display(Math('1.\\text{Summation Rule}')) 
display(Math("(f+g)' = f' + g' "))
display(Math("(f+g)' = %s"%(sym.latex(sym.diff(fplusg)))))
display(Math(" f' + g' = (%s)\\quad+\\quad (%s)\\quad =\\quad %s "\
             %(sym.latex(sym.diff(fx)),sym.latex(sym.diff(gx)),sym.latex(sym.expand(sym.diff(fx)+sym.diff(gx))))))
display(Math("\\text{Therefore: } (f+g)' = f' + g' "))
print('')   

#Product Rule 
display(Math('2.\\text{Product Rule}'))
display(Math("(f\\times g)' \\neq f' \\times g'"))
display(Math("(f\\times g)' = f'\\times g + f\\times g'"))
print('')
display(Math('\\text{Eg1. Without Product Rule}'))
display(Math(" %s\\quad \\neq\\quad %s "\
             %(sym.latex(sym.expand(sym.diff(fbyg))),sym.latex(sym.expand(sym.diff(fx)*sym.diff(gx))))))
print('')
display(Math('\\text{Eg2. With Product Rule}'))
display(Math("%s \\quad=\\quad %s "%(sym.latex(sym.expand(sym.diff(fbyg))),sym.latex(sym.expand(fdiv_xg + gdiv_xf)))))
```

```python
#DERIVATIVES OF TRIGONOMETRY FUNCTIONS 

#Showing the circular nature of the differential of sin and cos 
q = sym.symbols('q')

f = sym.cos(q)

for i in range(0,8): 
    display(Math("\\frac{df}{dx}%s = %s"%(sym.latex(f),sym.latex(sym.diff(f)))))
    f = sym.diff(f)
```

```python
#SHOWING THE ABOVE CODE AS AN IMAGE
import sympy.plotting.plot as symplot

f = sym.cos(q) 
clrs = 'krbg'
for i in range(0,4): 
    
    if i ==0: 
        p = symplot(f,show = False,label = f,line_color = clrs[i])
    else: 
        p.extend(symplot(f,show = False, label = f,line_color = clrs[i]))

    f = sym.diff(f)
p.legend = True 
p.xlim = [-3,3]
p.show()
```

```python
#DERIVATIVE OF TRIGONOMETRIC FUNCTIONS EXERCISE 
x,a = sym.symbols('x,a') 
f = sym.cos(x + sym.sin(x)) + a 
colors = 'brkm'

for ai in range(0,4): 
    if ai ==0: 
        p = symplot(f.subs(a,ai),show = False, label = 'a'+str(ai),line_color = colors[ai])
        
    else: 
        p.extend(symplot(f.subs(a,ai),show = False, label = 'a'+str(ai),line_color = colors[ai]))
p.title = 'the functions'
p.legend = True
p.show()

for ai in range(0,4): 
    if ai ==0: 
        p = symplot(sym.diff(f.subs(a,ai)),show = False, label = 'a'+str(ai),line_color = colors[ai])
        
    else: 
        p.extend(symplot(sym.diff(f.subs(a,ai)),show = False, label = 'a'+str(ai),line_color = colors[ai]))
        
p.title = 'Their derivatives' 
p.legend= True
p.show()
```
