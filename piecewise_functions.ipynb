```python
import numpy as np 
import sympy as sym 
import matplotlib.pyplot as plt 
from IPython.display import display, Math
import math
```

```python
from sympy.abc import x #assigns a symbolic variable to x, same as sym.symbol()

piece1 = 0 
piece2 = -2*x
piece3 = .1*x**3

fx = sym.Piecewise( (piece1, x<0),(piece2, (x>=0)&(x<3)),(piece3, x>=-3) )

fxx = sym.lambdify(x,fx)
xx = np.linspace(-3,5,1234)


plt.plot(xx,fxx(xx))
plt.show()
```

```python
#Exercise
from sympy.abc import x 

piece1 = x**3
piece2 = sym.log(x,2)

#Part 1: Printing out the function in latex
print('Part1: Printing out the function in Latex')
fx = sym.Piecewise( (piece1, x<=0),(piece2,x>0) )
display(Math('f(x) = ' + sym.latex(fx)))

print('')

#Part 2: Graphing the function
fxx = sym.lambdify(x,fx)
xx = np.linspace(-2,2,1234)
plt.plot(xx,fxx(xx),'k')
plt.title('Part2: Graphing the function')
plt.show()

#Part3: Graphing the function in xfkcd style 
with plt.xkcd():
    plt.plot(xx,fxx(xx),'k')
    plt.title('Part3: Function Grapphed in XKCD Style')
    plt.show()

```
